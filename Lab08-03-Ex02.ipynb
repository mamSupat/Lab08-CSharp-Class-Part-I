## คำถาม

1. Field ในภาษา C# คืออะไร


*คำตอบ*

 คือ ตัวแปรที่ประกาศภายในคลาสหรือโครงสร้าง (struct) ซึ่งใช้ในการเก็บข้อมูลที่เกี่ยวข้องกับอ็อบเจ็กต์หรือคลาสนั้นๆ โดยที่ Field จะถูกใช้ในการเก็บข้อมูลของอ็อบเจ็กต์และสามารถเข้าถึงหรือเปลี่ยนแปลงได้ตามที่กำหนดในโปรแกรม

2. Field มีหน้าที่อะไรใน class 


*คำตอบ*

เก็บข้อมูล ที่เกี่ยวข้องกับอ็อบเจ็กต์หรือคลาสนั้นๆ ซึ่งจะช่วยให้สามารถจัดเก็บสถานะหรือคุณสมบัติของอ็อบเจ็กต์ได้ ตัวแปรเหล่านี้จะเป็นตัวแสดงข้อมูลของอ็อบเจ็กต์และสามารถเปลี่ยนแปลงได้ตามที่กำหนดในโปรแกรม

3. การประกาศ Field ทำได้อย่างไร ให้อธิบายรูปแบบการประกาศ field ในคลาส


*คำตอบ*

กำหนดชื่อและประเภทของข้อมูล (Data Type) ที่ต้องการเก็บในคลาสนั้นๆ โดยมีรูปแบบดังนี้ : [AccessModifier] DataType FieldName [= InitialValue];


4. Field และ Variable แตกต่างกันอย่างไร


Field ใช้เก็บข้อมูลที่เกี่ยวข้องกับสถานะของอ็อบเจ็กต์หรือคลาส และมีอายุการใช้งานเท่ากับอ็อบเจ็กต์

Variable ใช้เก็บข้อมูลชั่วคราวในการทำงานภายในเมธอดหรือบล็อกเท่านั้น

5. เมื่อไหร่ควรใช้ Field และเมื่อไหร่ควรใช้ Variable

*คำตอบ*

ใช้ Field เมื่อข้อมูลนั้นต้องเก็บไว้ตลอดอายุของอ็อบเจ็กต์หรือคลาส และต้องการเข้าถึงจากหลายเมธอด

ใช้ Variable เมื่อข้อมูลนั้นชั่วคราวและใช้เฉพาะในระหว่างการทำงานของเมธอดหรือบล็อก
