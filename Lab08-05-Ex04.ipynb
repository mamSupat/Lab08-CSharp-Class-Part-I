## คำถาม

1. Field ในภาษา C# คืออะไร


*คำตอบ*

เป็นตัวแปรที่ประกาศภายในคลาสหรือสตรัค ซึ่งใช้เก็บข้อมูลของอ็อบเจ็กต์นั้น ๆ

2. Field มีหน้าที่อะไรใน class 


*คำตอบ*

1. ใช้เก็บข้อมูลของอ็อบเจ็กต์

2. กำหนดค่าเริ่มต้นให้กับอ็อบเจ็กต์

3. ใช้ร่วมกับ Constructor เพื่อกำหนดค่าเริ่มต้นแบบกำหนดเอง

4. ใช้ควบคุมการเข้าถึงข้อมูล

5. ใช้เก็บค่าคงที่และค่าคงอยู่ตลอดของคลาส

3. การประกาศ Field ทำได้อย่างไร

*คำตอบ*

ระบุ Access Modifier ระบุชนิดข้อมูลและตั้งชื่อ Field

4. Implicit fields initialization  คืออะไร

*คำตอบ*

ค่าดีฟอลต์ที่ C# กำหนดให้ Field ของคลาสโดยอัตโนมัติ เมื่อไม่มีการกำหนดค่าเริ่มต้นให้ Field นั้น ๆ

5. Explicit field initialization คืออะไร



*คำตอบ*

การกำหนดค่าให้กับ Field โดยตรงในขณะที่ประกาศ Field หรือผ่าน Constructor โดยผู้เขียนโค้ดจะกำหนดค่าที่ชัดเจนให้กับ Field เพื่อใช้ในการเริ่มต้นค่าเมื่ออ็อบเจ็กต์ถูกสร้างขึ้น

6. Implicit fields initialization  และ Explicit field initialization แตกต่างกันอย่างไร



*คำตอบ*

Implicit Field Initialization : C# จะกำหนดค่าเริ่มต้นให้กับ Field โดยอัตโนมัติ (เช่น 0, null ขึ้นอยู่กับประเภทข้อมูล)

Explicit Field Initialization : ผู้เขียนโค้ดกำหนดค่าเริ่มต้นให้กับ Field โดยตรงในการประกาศหรือผ่าน Constructor
